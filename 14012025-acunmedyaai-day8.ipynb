```python
#Sin ve Cos birlikte
x = np.linspace(0, 10, 2000)
plt.figure(figsize = (10, 6))
plt.plot(x, np.sin(x), label = 'Sin(x)')
plt.plot(x, np.cos(x), label = 'Cos(x)')
plt.xlabel('X', fontsize = 18)
plt.ylabel('Y', fontsize = 18)
plt.title('Sin(x) & Cos(x)', fontsize = 14)
plt.legend(loc = 'upper right')
plt.show()
```

```python
#Multiple graphs
plt.figure(figsize = (14, 6))
x = np.linspace(0, 10, 100)
y1 = np.sin(x) 
y2 = np.cos(x)
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.show()
```

```python
#Bar Graphs (Çubuk Grafikleri)
id1 = np.arange(1, 10)
score = np.arange(20, 110, 10)
plt.bar(id1, score)
plt.xlabel('Student ID')
plt.ylabel('Score')
plt.show()
```

```python
id1 = np.arange(1, 10)
score = np.arange(20, 110, 10)
plt.bar(id1, score, color = '#9A2EFE')
plt.xlabel('Student ID')
plt.ylabel('Score')
plt.show()
```

```python
#Multiple data
x1 = [1,3,5,7]
x2 = [2,4,6,8]
y1 = [7,7,7,7]
y2 = [17,18,29,40]
plt.figure(figsize = (8,6))
ax = plt.axes()
ax.set_facecolor("white")
plt.bar(x1, y1, label = 'First', color = '#42B300')
plt.bar(x2, y2, label = 'Second', color = '#94E413')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
Midterm1 = [28, 33, 43, 45, 57]
Name = ['Rasim','Fatih', 'Şeyma', 'Arda', 'Dilara']
plt.barh(Name, Midterm1, color = 'yellowgreen')
plt.show()
```

```python
num1 = np.array([1,3,5,7,9])
num2 = np.array([2,4,6,8,10])
plt.figure(figsize = (10, 6))
plt.bar(num1, num1**2, width = 0.3, color = '#FF6F00')
plt.bar(num2, num2**2, width = 0.3, color = '#FFB300')
for x, y in zip(num1, num1**2):
    plt.text(x, y + 0.05, '%d' % y, ha = 'center', va = 'bottom')
for x, y in zip(num2, num2**2):
    plt.text(x, y + 0.05, '%d' % y, ha = 'center', va = 'bottom')
plt.show()
```

```python
#stacked barplot

x1 = ['Abdülkadir', 'Cemile', 'Cihan', 'Dila']
y1 = [17, 18, 29, 40]
y2 = [20, 21, 22, 23]
plt.figure(figsize = (5,7)),
plt.bar(x1, y1, label = "Satışa Açık Biletler", width = 0.5, color = '#FF6F00')
plt.bar(x1, y2, label = "Satışa Kapalı Biletler", width = 0.5, bottom = y1, color = '#FFB300')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
x1 = ['Abdülkadir', 'Cemile', 'Cihan', 'Dila']
y1 = [17, 18, 29, 40]
y2 = [20, 21, 22, 23]

import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))
plt.barh(x1, y1, label="Satışa Açık Biletler", height=0.5, color="#FF6F00")
plt.barh(x1, y2, label="Satışa Kapalı Biletler", height=0.5, left=y1, color="#FFB300")
plt.xlabel('Y')
plt.ylabel('X')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
x1 = ['Abdülkadir', 'Cemile', 'Cihan', 'Dila']
y1 = [17, 18, 29, 40]
y2 = [20, 21, 22, 23]
plt.figure(figsize = (5,7)),
plt.bar(x1, y1, label = "Satışa Açık Biletler", width = 0.5, color = '#FF6F00')
plt.bar(x1, y2, label = "Satışa Kapalı Biletler", width = 0.5, bottom = y1, color = '#FFB300')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bar Chart')

for x, y in zip(x1, y1):
    plt.text(x, y - 10, '%d' % y, ha = 'center', va = 'bottom')
for x, y, z in zip(x1, y2, y1):
    plt.text(x, y + z - 10, '%d' % y, ha = 'center', va = 'bottom')

plt.legend()
plt.show()
```

```python
#Scatter Plots (Dağılım veya Saçılım Grafiği)
```

```python
x1 = np.array([250, 150, 350, 252, 450, 550, 455, 358, 158, 355])
y1 = np.array([40, 50, 80, 90, 100, 50, 60, 88, 54, 45])

#Graph-1
plt.scatter(x1, y1)
plt.xlabel('Time Spent', fontsize = 12)
plt.ylabel('Score', fontsize = 12)
plt.title('Scatter Graph')
plt.show()
```

```python
x2 = np.array([200, 100, 300, 220, 400, 500, 450, 380, 180, 350])
y2 = np.array([400, 500, 800, 900, 1000, 500, 600, 808, 504, 405])

#Graph-2
plt.scatter(x2, y2, color = 'r')
plt.xlabel('Time Spent', fontsize = 12)
plt.ylabel('Score', fontsize = 12)
plt.title('Scatter Graph')
plt.show()
```

```python
#Graph-3
plt.scatter(x1, y1, label = 'Class 1')
plt.scatter(x2, y2, label = 'Class 2', color = 'r')
plt.xlabel('Time Spent', fontsize = 12)
plt.ylabel('Score', fontsize = 12)
plt.title('Scatter Graph')
plt.legend()
plt.show()
```

```python
#Graph-4
plt.scatter(x1, y1, label = 'Class 1', marker = 'o', color = 'm')
plt.scatter(x2, y2, label = 'Class 2', marker = 'v', color = 'r')
plt.xlabel('Time Spent', fontsize = 12)
plt.ylabel('Score', fontsize = 12)
plt.title('Scatter Graph')
plt.legend()
plt.show()
```

```python
plt.scatter(np.random.random(10), np.random.random(10), c = 'r', s = 50, alpha = 0.6, label = 'One')
plt.scatter(np.random.random(10), np.random.random(10), c = 'b', s = 100, alpha = 0.6, label = 'Two')
plt.scatter(np.random.random(10), np.random.random(10), c = 'g', s = 150, alpha = 0.6, label = 'Three')
plt.scatter(np.random.random(10), np.random.random(10), c = 'y', s = 200, alpha = 0.6, label = 'Four')
plt.legend(bbox_to_anchor=(1.0, 1.0), fontsize = 'x-large')
plt.show()
```

```python
#Pie Charts (Pasta Grafikleri)
plt.figure(figsize = (9,9))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45)
plt.show()
```
