```python
import matplotlib.pyplot as plt
import neurolab as nl
```

```python
myinput = [[0, 0], [0, 1], [1, 0], [1, 1]]
target = [[0], [0], [0], [1]]
```

```python
#Tek elemanlı perceptron
net = nl.net.newp([[0, 1], [0, 1]], 1)
```

```python
error_progress = net.train(myinput, target, epochs = 100, show = 10, lr = 0.1) #learning rate = 0.1-0.0001
```

```python
plt.figure(figsize = (5,5))
plt.plot(error_progress)
plt.xlabel('Number of epochs')
plt.ylabel('Training Error')
plt.grid()
plt.show()
```

```python
#Tek katmanlı sinir ağı
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_data = np.loadtxt('/kaggle/input/neuralsimple/neural_simple.txt', delimiter = ',')
input_data
```

```python
#Verileri Ayrıştırma
data = input_data[:, 0:2]
labels = input_data[:, 2:]
```

```python
#Veriyi görselleştirme
plt.figure(figsize = (5,5))
plt.scatter(data[:, 0], data[:, 1])
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Input Data')
plt.show()
```

```python
dim1_min, dim1_max = data[:, 0].min(), data[:, 0].max()
dim2_min, dim2_max = data[:, 1].min(), data[:, 1].max()
```

```python
nn_output_layer = labels.shape[1]
nn_output_layer
```

```python
dim1 = [dim1_min, dim1_max]
dim2 = [dim2_min, dim2_max]
```

```python
neural_net = nl.net.newp([dim1, dim2], nn_output_layer)
```

```python
error = neural_net.train(data, labels, epochs = 200, show = 20, lr = 0.01)
```

```python
plt.figure()
plt.plot(error)
plt.xlabel('Number of epochs')
plt.ylabel('Training Error')
plt.title('Training Error Progress')
plt.grid()
plt.show()
```

```python
#test
print('\nTest Results:\n')
data_test = [[1.5, 3.2], [3.6, 1.7], [3.6, 5.7], [1.6, 3.9]]
for item in data_test:
    print(item, '-->', neural_net.sim([item])[0])
```
