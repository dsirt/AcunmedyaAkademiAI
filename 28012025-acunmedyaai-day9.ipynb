```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#pie chart
plt.figure(figsize = (9,9))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45)
plt.show()
```

```python
#Yüzdeleri gösterme
plt.figure(figsize = (8,8))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, autopct = '%1.1f%%')
plt.show()
```

```python
#hem yüzdelik değer hem de gerçek değer
plt.figure(figsize = (8,8))
area = [48, 30, 20, 15]
total = np.sum(area)
labels = ['Low', 'Medium', 'High', 'Critical']

def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, total*x/100)

colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, autopct = val_per)
plt.show()
```

```python
#explode
plt.figure(figsize = (8,8))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, autopct = '%1.1f%%', explode = [0, 0, 0, 0.1])
plt.show()
```

```python
#explode-2
plt.figure(figsize = (8,8))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, autopct = '%1.1f%%', explode = [0, 0, 0.1, 0.1])
plt.legend()
plt.show()
```

```python
#Donut Plot
plt.figure(figsize = (9,9))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
plt.figure(figsize = (9,9))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, autopct = '%1.1f%%', pctdistance = 0.85)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
plt.figure(figsize = (8,8))
area = [48, 30, 20, 15]
total = np.sum(area)
labels = ['Low', 'Medium', 'High', 'Critical']

def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, total*x/100)

colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, autopct = val_per, pctdistance = 0.85)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
#Multiple plot
fig = plt.figure(figsize = (20, 6))
area = [48, 30, 20, 15]
priority = ['Low', 'Medium', 'High', 'Critical']
status = ['Resolved', 'Cancelled', 'Pending', 'Assigned']
company = ['Anthropic', 'OpenAI', 'NVIDIA', 'DeepSeek']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']

plt.subplot(1,3,1)
plt.pie(area, labels = priority, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(1,3,2)
plt.pie(area, labels = status, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(1,3,3)
plt.pie(area, labels = company, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
#Histogram
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = "yellowgreen")
plt.gca().set(title = "Histogram", ylabel = "Frequency")
plt.show()
```

```python
#okunabilirliği arttırma
plt.figure(figsize = (10,8))
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = "yellowgreen", edgecolor = "#6A9662")
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
#multiple data
plt.figure(figsize = (8,11))
x = np.random.normal(-4, 1, size = 800)
y = np.random.normal(0, 1.5, size = 800)
z = np.random.normal(3.5, 1, size = 800)
plt.hist(x, bins = 30, color = 'yellowgreen', alpha = 0.6)
plt.hist(y, bins = 30, color = '#FF8F00', alpha = 0.6)
plt.hist(z, bins = 30, color = 'blue', alpha = 0.6)
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
col1 = np.linspace(0, 10, 1000)
col2 = np.sin(col1)
df = pd.DataFrame({'C1' : col1, 'C2' : col2})
df.head(10)
```

```python
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = df.C1, y = df.C2, data = df)
plt.show()
```

```python
#multiple data
plt.figure(figsize = (14, 6))
sns.set(rc = {"axes.facecolor" : "#FFFDE7", "axes.grid" : True, 'xtick.labelsize' : 14, 'ytick.labelsize' : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 2)
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 2)
plt.show()
```
